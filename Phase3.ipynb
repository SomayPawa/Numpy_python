# Advance operation with business example

```python
# data structure = [restaurant_id , 2021,2022,2023,2024]
import numpy as np
import matplotlib.pyplot as plt

sales_data = np.array([
    [1,150000,180000,220000,250000], # paradise Biryani
    [2,120000,140000,160000,190000], # beijing bites
    [3,200000,230000,260000,300000], # pizza hub
    [4,180000,210000,240000,270000], # burger point
    [5,160000,185000,205000,230000], # chai point
])

print("=== Zomato sales Analysis ===")

print("\n sales data shape ",sales_data.shape);


# ques-1 first three 3 restaurant data i have to print

print("first 3 res ", sales_data[:,:3]); # row , column : , :
print("first 3 res ", sales_data[:3]);
print("first 3 res ", sales_data[::3]);
print("first 3 res ", sales_data[:,1:]); # all row but after first column range hai 1 se start keroo end tak







```

## total sale per year


```python
# print("sales in 2021 ",sales_data[:,1:2]);
# print("sales in 2022 ",sales_data[:,2:3]);
# print("sales in 2023 ",sales_data[:,3:4]);
# print("sales in 2024 ",sales_data[:,4:5]);

print(np.sum(sales_data,axis=0));
yearly_tot = np.sum(sales_data[:,1:],axis=0);
print("yearly total",yearly_tot);

# minimum sales per restaurant

min_sales = np.min(sales_data[:,1:],axis=1);
print("minimum sales ",min_sales);

# maximum sales per year

max_sales_year = np.max(sales_data[:,1:],axis=0);
print("maximum sales per year ",max_sales_year);

# avg sales per restaurant 

avg_sales_avg = np.average(sales_data[:,1:],axis = 1); # axis = 0 means col , axis = 1 means row
avg_mean = np.mean(sales_data[:,1:],axis = 1); # axis = 0 means col , axis = 1 means row

print(" avg ",avg_sales_avg);
print(" mean ",avg_mean);





```

```python
# cumulative sale
cumsum = np.cumsum(sales_data[:,1:],axis=1);
print(cumsum);

plt.Figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("average cumulative sales per year across all the restaurant");
plt.xlabel("years");
plt.ylabel("sales");
plt.grid(True);
plt.show();
```
