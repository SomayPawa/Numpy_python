```python
# find the total number of missing values in each column

print(df.isnull().sum());
```

```python
## show all the record where AQI is greater then 300

df[df['AQI'] >= 300]
```

```python
## get all the row for the month of january 

df[df['Month'] == 1]
```

```python
## display the maximum PM2.5 VALUE and corresponding data 

df.iloc[0]
```

```python
## count how many days had holiday count has zero

cnt = df[df['Holidays_Count'] > 0]['Days'];
length = len(df[df['Holidays_Count'] > 0]);

print("cnt is",cnt);
print("length is ",length);
```

```python
## Find the average AQI for each month.

ans = df.groupby(['Month','Year'])['AQI'].mean();
print(ans);
```

```python
## Group the data by year and get the mean PM10 for each year.

ans = df.groupby('Year')['PM10'].mean();
print(ans);
```

```python
df.sort_values('PM2.5')
```

```python
## get the top 5 with highest ozone level
df.sort_values('Ozone');
ans = df.tail();
print(ans)
```

```python
## calculate the corelation 
# print(df);

corelation = df['PM2.5'].corr(df['AQI']);

print(corelation);
```

```python
## find the month which has highest average NO2
# print(df);

ans = df.groupby('Month')['NO2'].mean();

month = ans.idxmax();

print(month);

print(ans);
```

```python
## Replace all missing SO2 values with the mean SO2.

meanval = df['SO2'].mean();
print(meanval);

df['SO2'].fillna(meanval);



```
