```python
import matplotlib.pyplot as plt;
import numpy as np;
import pandas as pd;
```

### Matplot lib basic

```python
x = np.linspace(0,5,11);
y = x ** 2;
```

```python
plt.plot(x,y);
plt.title("x vs y")
plt.xlabel("x axis");
plt.ylabel("y axis");
```

```python
## sub plot -> multiple plot

plt.subplot(2,2,1) ## row and col ke hisab se divide ho jate hai (2*2) -> 2,2,3-> 2 * 2 main 3 i can select
plt.plot(x**2,x);
plt.subplot(2,2,4)
plt.plot(x,y);
plt.subplot(2,2,2)
plt.plot(y,x);
plt.subplot(2,2,3)
plt.plot(x**2,x**2);
```

```python
### object oriented way work 

fig = plt.figure(figsize=(6,6),dpi = 100)

axis1 = fig.add_axes([0.1,0.1,0.8,0.8]); ## left,bottom,width height

axis1.plot(x,y);

axis2 = fig.add_axes([0.15,0.45,0.4,0.4]); ## left,bottom,width height

axis2.plot(x,y);

```

```python
### type of plot

plt.scatter(x,y);



```

```python
plt.hist(x);
```

```python
plt.boxplot(x)
```

```python
### saving plot

fig = plt.figure()
ax = fig.add_axes([0.1,0.1,0.8,0.8])
ax.plot(x,y)
plt.savefig("basicplot.png");
```
